```python
import pandas as pd
import os
```

```python
os.listdir(path='arquivos')
```

```python
#Mesclando arquivos de dados de covid

df_covid = pd.DataFrame()
path='arquivos'
for arquivo in os.listdir(path):
     if arquivo.endswith(".csv"):
           df = pd.read_csv(f'{path}/{arquivo}', sep=";")
           df_covid = pd.concat([df_covid,df])

df_covid
```

```python
#Contando valores nulo em cada coluna
df_covid.isnull().sum()
```

```python
#avaliação do tipos de dados por coluna
df_covid.info()
```

```python
#Filtrando por colunas para análise
df_covid= df_covid[['regiao','estado','municipio','data','semanaEpi','casosAcumulado','casosNovos','obitosAcumulado','obitosNovos']]
```

```python
#convertendo coluna data para formato datetime
df_covid['data'] = pd.to_datetime(df_covid['data'])
```
